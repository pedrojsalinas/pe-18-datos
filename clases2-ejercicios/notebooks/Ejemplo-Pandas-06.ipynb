```python
mi_df['nombres-mayuscula'] = mi_df['nombres'].apply(lambda x: x.upper())
mi_df['apellidos-mayuscula'] = mi_df['apellidos'].apply(lambda x: x.upper())
```

```python
mi_df
```

```python
def count_nombres(x):
    valor = len(x)
    return valor
def count_apellidos(x):
    valor = len(x)
    if(valor>5):
        valor=5
    return valor
```

```python
mi_df['longitud-nombres'] = mi_df['nombres'].apply(count_nombres)
mi_df['longitud-apellidos'] = mi_df['apellidos'].apply(count_apellidos)
```
