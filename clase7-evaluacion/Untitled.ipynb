```python
import pandas as pd
import sys
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from pandas.tools.plotting import radviz

%matplotlib inline
```

```python
data = pd.read_csv('data/data-mundial-2018.csv', sep="|")
```

```python
data.DOB = pd.to_datetime(data.DOB)
data.DOB
```

```python
ax = data.groupby(['Club']).size().sort_values(ascending=False)[:20].plot.bar(title = u'Clubs', legend=True)

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
```

```python
data.set_index([data.ShirtName, data.Country, data.POS], drop=True,inplace=True)

ax = data[data['Caps']==1][['LastName', 'Caps']].sort_values(['Caps'], ascending=[False])\
.plot.bar(title = u'Jugadores con 0 partidos internacionales Fifa2018', legend=True, figsize=(16,5), stacked=True)

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```
