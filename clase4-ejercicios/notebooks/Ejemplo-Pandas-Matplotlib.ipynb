```python
pichincha = df[df['prov_insc'].str.strip()=='Pichincha'].groupby('dia_fall').prov_insc.count()
guayas = df[df['prov_insc'].str.strip()=='Loja'].groupby('dia_fall').prov_insc.count()
pichincha.plot(label='Pichincha')
guayas.plot(label='Loja')
plt.legend(loc='best')
```
